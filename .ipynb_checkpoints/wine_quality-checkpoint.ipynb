```python
#import necessary package
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import missingno as msno
import seaborn as sns
```

```python
#load dataset
dataset=pd.read_csv('wineQualityWhites.csv')
```

```python
dataset.head()
```

```python
#missing data detection
msno.matrix(dataset,figsize=(10,3))
```

```python
# Distribution of data
fig,axes=plt.subplots(nrows=2,ncols=1)
fig.set_size_inches(10,20)
sns.boxplot(data=dataset,orient='v',ax=axes[0])
sns.boxplot(data=dataset,orient='pH',y='quality',ax=axes[1])
plt.show()
```

```python
# correlation analysis
plt.figure(figsize=(15,10))
sns.heatmap(dataset.corr())
```

```python
# correlation analysis
corrMat=dataset.corr()
mask=np.array(corrMat)
mask[np.tril_indices_from(mask)]=False
fig,ax=plt.subplots()
fig.set_size_inches(20,10)
sns.heatmap(corrMat,mask=mask,square=True,annot=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
```

```python
# split the data into dependent and independent variable
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
# added extra column
x=np.append(arr=np.ones((x.shape[0],1)),values=x,axis=1)
```

```python
#split the dataset into train and test data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
# feature scalling
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
# train the model
lr=LinearRegression()
lr.fit(x_train,y_train)
predicts=lr.predict(x_test)
```

```python
# Evaluating and improve model
from sklearn.metrics import r2_score,mean_squared_error
```

```python
print(r2_score(y_test,predicts))
```
